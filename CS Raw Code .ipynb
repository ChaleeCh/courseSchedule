```python
# Here's the full code for the project
# importing libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

# loading the data 
def read_data(filepath):
    df = pd.read_csv(filepath)
    return df

df = read_data("C:/Datasets/Course Schedule.csv")
df

# drop columns we don't need
df = df.drop(columns= ["Timestamp", "WhatsApp Number"])
df

# make a copy of the original data
df1 = df.copy()
df1.head()

# renaming the columns 
df1 = df.rename(columns={'What are your available times on Saturday?': 'Saturday',
                        'What are your available times on Sunday?': 'Sunday'})
df1.head()

print(df1['Sunday'].value_counts())

'''
I'm mostly interested in Sunday because this is the day I initially set as the day for
which we will be having the class. So I'm going to proceed with the information in this column only.
'''

# create a sample dataframe based on the desired column
df_sunday = pd.DataFrame({'time_range': [
    '20hrs', '17:00', 'Any time after 6pm', '18-21hrs', '10 to 17',
    '6pm -8pm', '16:00 - 21:00', '8-9PM', '10:00 to 23:00', 'Yes',
    '8:00 PM', '15-16', 'Afternoon', '7-9 pm', '08 am to 14pm', '8-12',
    '20:00hr', '20:00', '8-10am', '14 till late'
]})

# define a function to categorize the time range
def categorize_time_range(time_range):
    if "am" in time_range or time_range.startswith("0"):
        return "morning-afternoon"
    elif "pm" in time_range or "-" in time_range:
        return "evening"
    elif "after" in time_range or "till" in time_range:
        return "evening"
    elif time_range == "Yes":
        return "evening"
    elif "Afternoon" in time_range:
        return "afternoon"
    else:
        return "evening-night"

# apply the categorize_time_range function to the time_range column
df_sunday['category'] = df_sunday['time_range'].apply(categorize_time_range)

# display the result
df_sunday

# check the exact count for each response 
df_sunday['category'].value_counts()

# visualize the results 
sns.countplot(x='category', data=df_sunday)
```
